#**KPI Requirements**

```python
#Total Revenue
total_revenue = df['total_price'].sum()

#Total Orders
total_unique_orders = df['order_id'].nunique()

#Average Order Value
avg_order_value = round(total_revenue/total_unique_orders,2)

#Total Pizza Sold
total_pizza_sold = df['quantity'].sum()

#Average Pizzas per Order
avg_pizzas_per_order = round(total_pizza_sold/total_unique_orders, 2)
```

```python
print(f"Total Revenue: {total_revenue}")
print(f"Total Orders: {total_unique_orders}")
print(f"Average Order Value: {avg_order_value}")
print(f"Total Pizza Sold: {total_pizza_sold}")
print(f"Average Pizzas Per Order: {avg_pizzas_per_order}")
```

##**Ingridient Analysis**

```python
ingredients = (
    df['pizza_ingredients'].str.split(',').explode().str.strip().value_counts().reset_index().rename(columns={'pizza_ingredients':'Ingredients',})
)
```

```python
print(ingredients.head(20))
```

```python
top_ingredients = ingredients.head(20).reset_index()
top_ingredients.columns = ['index', 'Ingredients', 'Count']

plt.figure(figsize=(12, 8))
sns.barplot(x='Ingredients', y='Count', data=top_ingredients, palette='viridis', hue='Ingredients', legend=False)
plt.title('Top 20 Most Frequent Pizza Ingredients')
plt.xlabel('Ingredients')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##**Daily Analysis**

```python
df['order_date']=pd.to_datetime(df['order_date'], dayfirst=True)
df['day_name'] = df['order_date'].dt.day_name()
week_day = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

df['day_name'] = pd.Categorical(df['day_name'], categories=week_day, ordered=True)
order_by_day = df.groupby('day_name',observed=False)['order_id'].nunique()
ax =  order_by_day.plot(kind='bar', figsize=(8,5), color='green', edgecolor='black')

plt.title("Total Orders by day of the week")
plt.xlabel("Day of Week")
plt.ylabel("Number of Orders")
plt.xticks(rotation=45)

for i, val in enumerate(order_by_day):
  plt.text(i, val+20, str(val), ha='center', va="bottom", fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```

#**Total Sales by day of Week bar plot**

```python


total_revenue_by_day = df.groupby('day_name', observed=False)['total_price'].sum()
pt = total_revenue_by_day.plot(kind='bar',figsize=(8,5), color='red', edgecolor='black')

plt.title('Total Sales by day of week')
plt.xlabel('Day of Week')
plt.ylabel('Total Sales Revenue')
plt.xticks(rotation=45)

for i, val in enumerate(total_revenue_by_day):
  plt.text(i, val+20, '$'+str(val),ha='center',va='bottom',fontsize=9,fontweight='bold')

plt.tight_layout()
plt.show()
```

#**Total Quantity Sold per day of Week**

```python


total_qty_sold_by_day = df.groupby('day_name', observed=False)['quantity'].sum()
pxt = total_qty_sold_by_day.plot(kind='bar', figsize=(8,5), color='purple', edgecolor='black')

plt.title('Total Quantity Sold by Day in Week')
plt.xlabel('Day in Week')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=45)

for i, val in enumerate(total_qty_sold_by_day):
  plt.text(i, val+20, str(val), va='bottom', ha='center',fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```

##**Hourly Trends**

```python
df['order_time'] = pd.to_datetime(df['order_time'], format='%H:%M:%S')
df['order_hour'] = df['order_time'].dt.hour

order_by_hour = df.groupby('order_hour', observed=False)['order_id'].nunique()
ax = order_by_hour.plot(kind='bar', figsize=(8,5), color="maroon")

plt.xlabel('Hour in (24-hr-format)')
plt.ylabel('Total Orders Placed')
plt.xticks(rotation=45)
plt.title('Total Orders Place By Each Hour')

for i, val in enumerate(order_by_hour):
  plt.text(i, val+20, str(val), ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()
```

# **Total Order Placed By months**

```python
df['month_name'] = df['order_date'].dt.month_name();
months_in_year = ['January', 'February', 'March', 'April', "May", 'June', 'July','August', 'September',
                  'October', 'November', 'December']

df['month_name'] = pd.Categorical(df['month_name'], categories=months_in_year, ordered=True)
orders_placed_by_months = df.groupby('month_name', observed=False)['order_id'].nunique()

plt.figure(figsize=(8, 5))
plt.fill_between(orders_placed_by_months.index, orders_placed_by_months.values, color='orange', alpha=0.6)
plt.plot(orders_placed_by_months.index, orders_placed_by_months.values, color='black', linewidth=2, marker='o')


plt.title('Total Orders Place By Month')
plt.xlabel('Months in Year')
plt.ylabel('Total Orders Placed')
plt.xticks(rotation=45)

for month, count in enumerate(orders_placed_by_months):
  plt.text(month, count, str(count), ha='center', va='bottom', fontweight='bold', fontsize=9)

plt.tight_layout()
plt.show()

```

# **Percentage of Sales By Pizza Category**

```python


pizza_category_sales = df.groupby('pizza_category', observed = True)['total_price'].sum()
category_percent = pizza_category_sales/pizza_category_sales.sum() * 100

plt.figure(figsize=(5,5))
colors = plt.get_cmap('tab20').colors
plt.pie(category_percent, labels=category_percent.index, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops={'edgecolor':'black', 'width':0.4})
plt.title("Percentage of Sales by Pizza Category")

plt.show()

```

# **Percentage of Sales by Size and Category**

```python
sales_pivot = df.pivot_table(
    index='pizza_category',
    columns='pizza_size',
    values= 'total_price',
    aggfunc = 'sum',
    fill_value = 0
)

sales_percent = sales_pivot / sales_pivot.sum().sum() * 100

plt.figure(figsize=(10,6))
sns.heatmap(sales_percent, annot=True, fmt='.1f', cmap='YlOrRd', linewidth=0.5)
plt.title('Percentage of Sales by Size and Category')
plt.xlabel('Pizza Size')
plt.ylabel('Pizza Category')
plt.show()
```

# **Total Pizza Sold by Pizza Category**

```python
total_pizza_sold_by_pizza_category = df.groupby('pizza_category', observed=False)['quantity'].sum()
colors= list(plt.get_cmap('tab20').colors)
colors=colors[:len(total_pizza_sold_by_pizza_category)]

total_pizza_sold_by_pizza_category.plot(kind='bar', color=colors, figsize=(7,5), edgecolor='black')

plt.title("Total Pizza Sold by Pizza Category", fontdict= {'fontsize': 16, 'fontweight':'bold'})
plt.xlabel('Pizza Category')
plt.ylabel('Total Pizza Sold')
plt.xticks(rotation=45)

for cat, quantity in enumerate(total_pizza_sold_by_pizza_category):
  plt.text(cat, quantity, str(quantity), fontsize=9, va='bottom', ha='center',fontweight='bold')


plt.tight_layout()
plt.show()
```

# **Top 5 Best Selling Pizzas - Total Quantity**

```python
pizza_name_by_quantity_sold = df.groupby('pizza_name', observed=False)['quantity'].sum()
top5 = pizza_name_by_quantity_sold.sort_values(ascending=False).head(5)

colours = list(plt.get_cmap('tab20').colors)
colours= colours[:len(top5)]

top5.plot(kind='bar', color=colours, figsize=(7,5), edgecolor='black')
plt.ylim(0, 3000)

plt.title('Top 5 Pizza by Total quantity sold', fontdict={'fontsize':16, 'fontweight':'bold'})
plt.xlabel('Pizza Name')
plt.ylabel('Total Quantity Sold')
plt.xticks(rotation=38)


for name, quantity in enumerate(top5):
  plt.text(name, quantity, str(quantity), ha='center', va='bottom', fontdict={'fontsize':9, 'fontweight':'bold'})

plt.tight_layout()
plt.show()
```

# **Top 5 Best Selling Pizzas - Total Orders**

```python
pizza_sold_by_total_orders = df.groupby('pizza_name', observed=False)['order_id'].nunique()
top5 = pizza_sold_by_total_orders.sort_values(ascending = False).head(5)


top5.plot(kind='bar', color='orange', figsize=(6,5), edgecolor='black', alpha=0.7)
plt.ylim(0, 3000)
plt.title('TOP 5 PIZZA BY TOTAL ORDERS')
plt.xlabel('Pizza Name')
plt.ylabel('Total Orders Placed')
plt.xticks(rotation=45)

for name, order_count in enumerate(top5):
  plt.text(name, order_count, str(order_count), fontsize=9, fontweight='bold', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

# **Top 5 Best Selling Pizzas-Total Revenue**

```python
total_revenue_by_pizzas = df.groupby('pizza_name', observed=False)['total_price'].sum()
top5 = total_revenue_by_pizzas.sort_values(ascending=False).head(5)

top5.plot(kind= 'bar', color='brown', alpha=0.6, edgecolor = 'black', figsize=(8, 6))
plt.ylim(0, 50000)
plt.title('Top 5 Best Selling Pizzas by Total Revenue')
plt.xlabel('pizza name')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)

for name, total in enumerate(top5):
  plt.text(name, total, str(total), fontweight='bold', fontsize=9, ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
